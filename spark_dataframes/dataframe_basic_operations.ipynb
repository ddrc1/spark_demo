```python
import os
import findspark
findspark.init(os.getenv('SPARK_HOME'))
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("ops").getOrCreate()
```

```python
df = spark.read.csv("./appl_stock.csv", header=True, inferSchema=True)
df.show()
```

```python
df.printSchema()
```

```python
filter_df = df.filter('Close > 600')
filter_df.show()
```

```python
filter_df.select(['Open', 'Close']).show()
```

```python
df.filter(df['Close'] > 600).show()
```

```python
df.filter((df['Close'] < 200) & (df['Close'] > 100)).show()
```

```python
df.filter(df['Low'] == 197.16).show()
```

```python
result = df.filter(df['Low'] == 197.16).collect()
result
```
